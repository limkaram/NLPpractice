## 코사인 유사도(Cosine Similarity)

* 두 벡터간 코사인 각도를 활용하여 두 벡터의 유사한 정도를 파악
* 두 벡터의 방향이 완전히 동일한 경우 1의 값을 가지고, 방향이 완전히 반대라면 -1의 값을 가짐
* 즉 -1이상 1이하의 값을 가지며, 1에 가까울 수록 유사도가 높다고 판단

스크린샷 2020-11-07 오후 3.20.05.png

```
import numpy as np

def cos_sim(A, B):
    # np.dot(A, B) : 벡터 A와 B의 내적 값 반환
    # np.norm(A) : 벡터 A의 길이 반환(L2 norm)
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))

# BoW 예시
doc1=np.array([0,1,1,1])
doc2=np.array([1,0,1,1])
doc3=np.array([2,0,2,2])

print(cos_sim(doc1, doc2))
print(cos_sim(doc1, doc3))
print(cos_sim(doc2, doc3))
```
